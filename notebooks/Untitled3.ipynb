```python
# Step 1: Load cleaned data and add date-based features
import pandas as pd
from pathlib import Path

# Load your saved dataset
clean_path = Path("kwh_prediction/data/clean/energy_clean_no_payment.parquet")
df = pd.read_parquet(clean_path)
print("Loaded:", df.shape)
print(df.head(3))

# --- Date-based features ---
df["year"]       = df["chargedate"].dt.year
df["month_num"]  = df["chargedate"].dt.month
df["quarter"]    = df["chargedate"].dt.quarter
df["dayofyear"]  = df["chargedate"].dt.dayofyear

# Simple season mapping (adjust for your region if needed)
def get_season(m):
    if m in [12, 1, 2]:  return "winter"
    elif m in [3, 4, 5]: return "spring"
    elif m in [6, 7, 8]: return "summer"
    else:                return "fall"

df["season"] = df["month_num"].apply(get_season)

print("\nNew time columns:", ["year","month_num","quarter","dayofyear","season"])
display(df.head(5))
```
